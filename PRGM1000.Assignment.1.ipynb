## Development Platform

In this course, we will be using the latest version of Python 3 (Python 3.7.4, as this is being written) as a scripting language for building scripts and tools for network and system management. (Later, we will also look at Bash and Powershell, but even then it will be in the context of comparing them with Python).

We will also use the suite of standard tools provided by the [Anaconda](https://www.anaconda.com) project. In addition to providing the Python3 interpreter and toolchain, Anaconda will supply us with the [Spyder](https://spyder-ide.org) integrated development environment and the [Jupyter](https://jupyter.org) platform for interactive computing. You will become quite familiar with these tools over the duration of this course. Python projects have a noticeable fetish for having the diagram "py" in their names.

### 4 - Save and Upload Your Work

One command you should be familiar with if you're going to use Jupyter notebooks to complete assignments is <span style="color:blue">"Save and Checkpoint"</span>, which will capture the current state of your notebook on the server so that you can revert to it if you make a mistake or if the server crashes or stops for some reason.



Now, in order to get credit for this assignment, you need to be able to capture it's current state.  This is non-trivial, because the current "state" of the notebook only exists on your Jupyter server.  In order to push it down to a file, you need to:

1. Use the "Save" or "Save as" commands in the File menu to make sure the current state is tagged, then:
2. Use the "Download" command to send it from the server to your hard-drive as a file.

In general, you'll want either "notebook" (for a copy of the notebook that can be uploaded and edited) or "PDF" for a printable fixed view of the current state.
